```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("pyspark_5").getOrCreate()
```

```python
data1 = [
    ("Mark",1000),
    ("Bob",1500),
    ("Charlie",None),
    ("David",2200),
    ("Eve",1200),
]
```

```python
# create dataframe from data1
df = spark.createDataFrame(data1, ["name", "salary"])

#filter rows with null
df.filter(df.salary.isNull()).show()
```

```python
##filter rows with  not null
df.filter(df.salary.isNotNull()).show()
```

```python
#Replacing null with 0 for Data cleaning
df_clean = df.fillna(0)
df_clean.show()
```

```python
#combining isNull() and isNotNull() with Other Filters

df.filter(df.salary.isNull()).filter(df.name.isNotNull()).show()

```

```python
#combining isNull() and isNotNull() with Other Filters

df.filter(df.salary.isNotNull() & (df.salary > 1000)).show()
```
